```python
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import joblib
```

```python
df = pd.read_csv("/content/data_credito.csv", sep=';', decimal=',')
df.drop(columns=['ID_Cliente'], inplace=True, errors='ignore')  # Eliminamos ID si existe
```

```python
df['Estado_Civil'].unique()
```

```python
# Definir X e y
X = df.drop(columns=["Incumplimiento"])
y = df["Incumplimiento"]

# Detectar tipos de variables
num_cols = X.select_dtypes(include=["int64", "float64"]).columns.tolist()
cat_cols = X.select_dtypes(include=["object", "category"]).columns.tolist()
```

```python
# Transformadores
numeric_transformer = Pipeline([
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline([
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Composición del preprocesamiento
preprocessor = ColumnTransformer([
    ('num', numeric_transformer, num_cols),
    ('cat', categorical_transformer, cat_cols)
])

# Pipeline final
pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression(max_iter=1000))
])

# División de datos
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)

# Entrenamiento
pipeline.fit(X_train, y_train)
```

```python
joblib.dump(pipeline, "modelo_crediticio.pkl")
```
