```python
# Importación de librerías necesarias para el proyecto
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, roc_auc_score, silhouette_score
from sklearn.impute import SimpleImputer
import joblib
import os
import warnings
from sklearn.exceptions import ConvergenceWarning
import pickle


#====================================================================================================
# Cargar el conjunto de datos
try:
    # Para Google Colab, es necesario subir el archivo
    from google.colab import files
    uploaded = files.upload()
    data = pd.read_csv('/content/data_credito.csv', sep=';', decimal=',')
except:
    # Alternativa si el archivo ya está en el entorno
    data = pd.read_csv('/content/data_credito.csv', sep=';', decimal=',')

# Eliminar 'ID_Cliente' después de cargar
data.drop(columns=['ID_Cliente'], inplace=True)

# Mostrar las primeras filas del dataset
print("\nMuestra de los datos:")
print(data.head())

# Información general del dataset
print("\nInformación del dataset:")
print(data.info())

# Estadísticas descriptivas
print("\nEstadísticas descriptivas:")
print(data.describe())

# Verificar valores faltantes
print("\nValores faltantes por columna:")
print(data.isnull().sum())


# FunciónAnálisis univariado para variables numéricas
def analisis_univariado_numerico():
    # Seleccionar variables numéricas
    variables_numericas = data.select_dtypes(include=['int64', 'float64']).columns.tolist()
    variables_numericas.remove('ID_Cliente') if 'ID_Cliente' in variables_numericas else None

    # Crear subplots para histogramas y boxplots
    n_variables = len(variables_numericas)
    fig, axes = plt.subplots(n_variables, 2, figsize=(15, 4*n_variables))

    for i, var in enumerate(variables_numericas):
        # Histograma
        sns.histplot(data[var], kde=True, ax=axes[i, 0])
        axes[i, 0].set_title(f'Distribución de {var}')

        # Boxplot
        sns.boxplot(x=data[var], ax=axes[i, 1])
        axes[i, 1].set_title(f'Boxplot de {var}')

    plt.tight_layout()
    plt.show()
#=====================================================================================================
# Ejecutar análisis univariado variables numéricas
analisis_univariado_numerico()

# 1. Seleccionar variables numéricas y categóricas
numeric_cols = data.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = data.select_dtypes(include=['object', 'category']).columns

# 2. Aplicar one-hot encoding a variables categóricas
data_encoded = pd.get_dummies(data, columns=categorical_cols, drop_first=True)

# 3. Calcular matriz de correlación (incluyendo las dummies)
corr_matrix = data_encoded.corr(method='pearson')
#print(corr_matrix)
# 4. Visualizar con heatmap (escala fija -1 a 1)
plt.figure(figsize=(16, 12))
sns.heatmap(
    corr_matrix,
    annot=True,
    fmt=".2f",
    cmap='coolwarm',
    center=0,
    vmin=-1,
    vmax=1,
    mask=np.triu(np.ones_like(corr_matrix, dtype=bool))  # Máscara para triangular superior
)
plt.title("Matriz de Correlación (Incluyendo Variables Categóricas Codificadas)")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Filtrar correlaciones absolutas > 0.5
strong_corr = corr_matrix[(corr_matrix.abs() > 0.5) & (corr_matrix != 1.0)].stack()
print(strong_corr)

# Seleccionar solo variables numéricas
numeric_cols = data.select_dtypes(include=['int64', 'float64']).columns
corr_matrix = data[numeric_cols].corr(method='pearson')  # Correlación de Pearson
#print(corr_matrix)
# Visualizar con heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', center=0, mask=np.triu(corr_matrix),vmin=-1,vmax=1)
plt.title("Matriz de Correlación (Variables Numéricas Crudas)")
plt.show()

# Filtrar correlaciones con 'Incumplimiento' y ordenar
corr_with_target = corr_matrix['Incumplimiento'].sort_values(key=abs, ascending=False)
print(corr_with_target)

# Filtrar variables con correlación significativa con el target
significant_vars = ['Historial_Crediticio', 'Tiempo_Meses_Prestamo', 'Deuda_Actual', 'Incumplimiento']
sns.heatmap(corr_matrix.loc[significant_vars, significant_vars], annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Correlaciones Relevantes con Incumplimiento")
plt.show()

# Pairplot para relaciones visuales (muestra solo 4 variables para claridad)
sns.pairplot(data[['Ingreso_Mensual', 'Deuda_Actual', 'Monto_Prestamo', 'Historial_Crediticio']])
plt.suptitle("Relaciones Crudas entre Variables Numéricas", y=1.02)
plt.show()

# Suponiendo que 'Incumplimiento' es numérico (0 y 1)
target_corr = data[numeric_cols].corrwith(data['Incumplimiento']).sort_values(ascending=False)

# Gráfico de barras
plt.figure(figsize=(10, 6))
target_corr.drop('Incumplimiento').plot.barh(color='skyblue')
#print(target_corr)
plt.title("Correlación Cruda con el Target (Incumplimiento)")
plt.xlabel("Coeficiente de Correlación")
plt.grid(axis='x', linestyle='--')
plt.show()

# Funciom Análisis univariado para variables categóricas
def analisis_univariado_categorico():
    # Seleccionar variables categóricas
    variables_categoricas = data.select_dtypes(include=['object']).columns.tolist()

    # Crear subplots para gráficos de barras y countplots
    n_variables = len(variables_categoricas)
    fig, axes = plt.subplots(n_variables, 1, figsize=(12, 4*n_variables))

    # Si solo hay una variable categórica, aseguramos que axes sea iterable
    if n_variables == 1:
        axes = [axes]

    for i, var in enumerate(variables_categoricas):
        # Gráfico de barras con porcentajes
        counts = data[var].value_counts()
        percentages = 100 * counts / len(data)

        sns.barplot(x=percentages.index, y=percentages.values, ax=axes[i])
        axes[i].set_title(f'Distribución de {var} (%)')
        axes[i].set_ylabel('Porcentaje')

        # Añadir etiquetas de porcentaje
        for j, p in enumerate(percentages):
            axes[i].annotate(f'{p:.1f}%',
                             (j, p),
                             ha='center',
                             va='bottom')

    plt.tight_layout()
    plt.show()


# Ejecutar análisis univariado variables numéricas
analisis_univariado_categorico()



# Análisis de la variable objetivo
plt.figure(figsize=(10, 6))
sns.countplot(x='Incumplimiento', data=data)
plt.title('Distribución de la Variable Objetivo: Incumplimiento')
counts = data['Incumplimiento'].value_counts()
for i, count in enumerate(counts):
    plt.annotate(f'{count} ({100 * count / len(data):.1f}%)',
                 (i, count),
                 ha='center',
                 va='bottom')
plt.show()

#==============================================================================================
# Convertir variables categóricas a numéricas (ejemplo básico)
data = pd.get_dummies(data, columns=['Tipo_Empleo', 'Estado_Civil', 'Sexo'], drop_first=True)

# Separar features y target
X = data.drop('Incumplimiento', axis=1)
y = data['Incumplimiento']

# 2. División en train/test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Escalar variables numéricas (opcional para RF, pero útil para comparar importancias)
numeric_cols = ['Edad', 'Ingreso_Mensual', 'Historial_Crediticio', 'Deuda_Actual',
                'Tiempo_Empleo', 'Numero_Productos', 'Monto_Prestamo', 'Tiempo_Meses_Prestamo']
scaler = StandardScaler()
X_train[numeric_cols] = scaler.fit_transform(X_train[numeric_cols])
X_test[numeric_cols] = scaler.transform(X_test[numeric_cols])

# 3. Entrenamiento del modelo Random Forest
rf = RandomForestClassifier(
    n_estimators=200,
    max_depth=10,
    min_samples_split=5,
    min_samples_leaf=2,
    max_features='sqrt',
    random_state=42,
    class_weight='balanced'  # Para manejar desbalance de clases
)

rf.fit(X_train, y_train)

# 4. Evaluación del modelo
y_pred = rf.predict(X_test)
y_prob = rf.predict_proba(X_test)[:, 1]

print("=== Métricas de Evaluación ===")
print(classification_report(y_test, y_pred))
print("\nMatriz de Confusión:")
print(confusion_matrix(y_test, y_pred))
print(f"\nAUC-ROC: {roc_auc_score(y_test, y_prob):.4f}")

# 5. Análisis de importancia de variables
importances = rf.feature_importances_
features = X.columns
indices = np.argsort(importances)[::-1]

print("\n=== Importancia de Variables ===")
for f in range(X_train.shape[1]):
    print(f"{features[indices[f]]}: {importances[indices[f]]:.4f}")

# Visualización
plt.figure(figsize=(12, 8))
plt.title("Importancia de Variables - Random Forest")
sns.barplot(x=importances[indices], y=features[indices])
plt.tight_layout()
plt.show()

#========================================================================================
df = data.copy()

# 1. Tratamiento de Outliers: Transformación logarítmica
df['Ingreso_Mensual_log'] = np.log1p(df['Ingreso_Mensual'])

# 2. Creación del Índice de Incumplimiento (usando Ingreso_Mensual_log)
variables = [
    'Historial_Crediticio',
    'Deuda_Actual',
    'Ingreso_Mensual_log',  # Usamos la versión transformada
    'Monto_Prestamo',
    'Tiempo_Meses_Prestamo'
]
pesos = [0.30, 0.20, 0.15, 0.20, 0.15]

# Normalización de variables (StandardScaler)
scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df[variables]), columns=variables)

# Cálculo del índice ponderado
df['Indice_Incumplimiento'] = (
    df_scaled['Historial_Crediticio'] * pesos[0] +
    df_scaled['Deuda_Actual'] * pesos[1] +
    df_scaled['Ingreso_Mensual_log'] * pesos[2] +  # Peso aplicado a la versión transformada
    df_scaled['Monto_Prestamo'] * pesos[3] +
    df_scaled['Tiempo_Meses_Prestamo'] * pesos[4]
)

# 3. Ingeniería de Características Adicionales (opcional)
df['Ratio_Deuda_Ingreso'] = df['Deuda_Actual'] / (df['Ingreso_Mensual'] + 1e-6)  # Usar Ingreso_Mensual original para interpretabilidad
df['Edad_Bin'] = pd.cut(df['Edad'], bins=[18, 30, 45, 60, 75, 90], labels=['18-30', '31-45', '46-60', '61-75', '76+'])

# 4. Validación
correlation = df[['Indice_Incumplimiento', 'Incumplimiento']].corr().iloc[0,1]
print(f"Correlación entre Índice e Incumplimiento: {correlation:.2f}")

# Visualización del índice
plt.figure(figsize=(10, 4))
sns.histplot(df['Indice_Incumplimiento'], kde=True, bins=30)
plt.title('Distribución del Índice de Incumplimiento (usando Ingreso_Mensual_log)')
plt.show()

# Muestra de resultados
print("\nDataFrame procesado (5 primeras filas):")
print(df[['Ingreso_Mensual', 'Ingreso_Mensual_log', 'Indice_Incumplimiento', 'Ratio_Deuda_Ingreso', 'Edad_Bin']].head())

df['Historial_Crediticio_Bin'] = pd.cut(df['Historial_Crediticio'], bins=[300,400,600,800,850],labels=['Bajo','Regular','Bueno','Excelente'])
df['Ingreso_Mensual_Bin']=pd.cut(df['Ingreso_Mensual_log'],bins=[6.4,7.5,9.0,20.0], labels=['Bajo','Medio','Alto'])
df['Deuda_Actual_Bin']=pd.cut(df['Deuda_Actual'],bins=[0,18000,35000,50000],labels=['Bajo','Medio','Alto'])
df['Plazo']=pd.cut(df['Tiempo_Meses_Prestamo'],bins=[0,12,36,60],labels=['Corto_Plazo','Mediano_Plazo','Largo_Plazo'])

def categorizar_tiempo_empleo(meses):
    if meses == 0:
        return "Desempleado"
    elif 1 <= meses <= 12:
        return "1 a 12 meses"
    elif 13 <= meses <= 60:
        return "1 a 5 años"
    elif 61 <= meses <= 120:
        return "5 a 10 años"
    elif meses > 120:
        return "Más de 10 años"
    else:
        return np.nan  # Para valores negativos o nulos

# Aplicar la función a la columna 'tiempo_empleo'
df['Tiempo_Empleo_Bin'] = df['Tiempo_Empleo'].apply(categorizar_tiempo_empleo)


# --- Función auxiliar para aplicar get_dummies de forma segura ---
def apply_dummies_safely(dataframe, column_name, prefix, drop_original=True):
    """
    Aplica one-hot encoding a una columna, concatena los resultados
    y opcionalmente elimina la columna original.
    Maneja el caso donde la columna no existe.
    """
    if column_name not in dataframe.columns:
        print(f"Advertencia: La columna '{column_name}' no existe en el DataFrame. No se crearán dummies.")
        return dataframe

    # `dummy_na=False` es el valor por defecto, no crea una columna para NaNs.
    # Si quisieras una columna explícita para NaN, usa dummy_na=True.
    dummies = pd.get_dummies(dataframe[column_name], prefix=prefix, dtype=int) # dtype=int para asegurar 0/1 enteros

    # Concatenar los dummies. Si hay nombres de columnas que ya existen
    # (por ejemplo, si el prefijo + categoría ya existe, lo cual es raro
    # pero podría pasar con un mal prefijo), pd.concat los manejará.
    # La validación de duplicados la haremos al final.
    dataframe = pd.concat([dataframe, dummies], axis=1)

    if drop_original:
        dataframe = dataframe.drop(columns=[column_name])
        print(f"Columna original '{column_name}' eliminada.")

    return dataframe

# --- Tu script para Dicotomizar (One-Hot Encoding) con la función segura ---

# 1. Dicotomizar Edad_Bin
df = apply_dummies_safely(df, 'Edad_Bin', 'Edad')

# 2. Dicotomizar Historial_Crediticio_Bin
df = apply_dummies_safely(df, 'Historial_Crediticio_Bin', 'Historial')

# 3. Dicotomizar Ingreso_Mensual_Bin
df = apply_dummies_safely(df, 'Ingreso_Mensual_Bin', 'Ingreso')

# 4. Dicotomizar Deuda_Actual_Bin
df = apply_dummies_safely(df, 'Deuda_Actual_Bin', 'Deuda')

# 5. Dicotomizar Tiempo_Empleo_Bin
# ¡Esta es la ÚNICA VEZ que se procesa 'Tiempo_Empleo_Bin'!
df = apply_dummies_safely(df, 'Tiempo_Empleo_Bin', 'Tiempo')

# 6. Dicotomizar Plazo
df = apply_dummies_safely(df, 'Plazo', 'Plazo')
#df = df.drop(columns=columnas_eliminar)
#tipos_a_excluir = ['category', 'object']
#df = df.select_dtypes(exclude=tipos_a_excluir)

columnsSelected = ['Incumplimiento', 'Tipo_Empleo_Empleado',
       'Tipo_Empleo_Estudiante', 'Tipo_Empleo_Independiente',
       'Estado_Civil_Divorciado', 'Estado_Civil_Soltero', 'Estado_Civil_Viudo',
       'Sexo_Masculino', 'Indice_Incumplimiento',
       'Ratio_Deuda_Ingreso', 'Edad_18-30', 'Edad_31-45', 'Edad_46-60',
       'Edad_61-75', 'Edad_76+', 'Historial_Bajo', 'Historial_Regular',
       'Historial_Bueno', 'Historial_Excelente', 'Ingreso_Bajo',
       'Ingreso_Medio', 'Ingreso_Alto', 'Deuda_Bajo', 'Deuda_Medio',
       'Deuda_Alto', 'Tiempo_1 a 12 meses', 'Tiempo_1 a 5 años',
       'Tiempo_5 a 10 años', 'Tiempo_Desempleado', 'Tiempo_Más de 10 años',
       'Plazo_Corto_Plazo', 'Plazo_Mediano_Plazo', 'Plazo_Largo_Plazo']

Columnuique= list(pd.Series(columnsSelected).drop_duplicates())
df=df[Columnuique]


from scipy.stats import ttest_ind, chi2_contingency

# Asegúrate de que las columnas booleanas sean tratadas como ints para Chi-cuadrado
for col in df.select_dtypes(include='bool').columns:
    df[col] = df[col].astype(int)
# --- FIN DE LA SIMULACIÓN ---


# 1. Definir la variable objetivo
target_variable = 'Incumplimiento'

# 2. Separar variables numéricas y categóricas (dicotomizadas/dummies)
# Basado en tu df.info()
numeric_cols = ['Indice_Incumplimiento', 'Ratio_Deuda_Ingreso']

# Todas las demás columnas que son int64 y no son la target son dummies.
# Las columnas 'bool' también se tratarán como categóricas/dummies (0/1).
categorical_cols = [col for col in df.columns if col not in numeric_cols + [target_variable]]

print(f"\nVariables Numéricas para análisis bivariado: {numeric_cols}")
print(f"Variables Categóricas (incluyendo dummies) para análisis bivariado: {categorical_cols}")



## Análisis para Variables Categóricas (Dicotomizadas/Dummie

results_categorical = []

print("\n" + "="*70)
print("INICIANDO ANÁLISIS BIVARIADO: VARIABLES CATEGÓRICAS vs. INCUMPLIMIENTO")
print("="*70 + "\n")

for col in categorical_cols:
    print(f"\n--- Analizando: {col} ---")

    # Asegúrate de que la columna no esté vacía o tenga demasiados NaNs
    if df[col].isnull().all():
        print(f"Advertencia: La columna '{col}' contiene solo valores nulos. Saltando análisis.")
        continue

    # Asegura que la columna sea numérica (0/1) para los cálculos, si es bool
    if df[col].dtype == 'bool':
        df[col] = df[col].astype(int)

    # 1. Tabla de contingencia (porcentajes por columna de Incumplimiento)
    # pd.crosstab es robusto y maneja NaNs excluyéndolos por defecto.
    contingency_table = pd.crosstab(df[col], df[target_variable], normalize='columns') * 100
    print("\nTabla de Porcentajes de Incumplimiento por categoría:")
    print(contingency_table)

    # 2. Chi-cuadrado para independencia
    # Asegúrate de que la tabla de contingencia no esté vacía y tenga al menos 2x2.
    # El test de chi-cuadrado requiere que haya más de una categoría en ambas variables.
    if contingency_table.shape[0] > 1 and contingency_table.shape[1] > 1 and not contingency_table.empty:
        # Asegúrate de pasar la tabla de conteos, no de porcentajes, a chi2_contingency
        chi2, p_val, dof, expected = chi2_contingency(pd.crosstab(df[col], df[target_variable]))
        print(f"\nTest de Chi-cuadrado - p-value: {p_val:.4f}")
        if p_val < 0.05:
            print(f"  -> Hay una relación estadísticamente significativa entre {col} y {target_variable} (p < 0.05).")
        else:
            print(f"  -> No hay evidencia de una relación estadísticamente significativa entre {col} y {target_variable} (p >= 0.05).")
    else:
        print("\nNota: Chi-cuadrado no aplicable o no tiene suficiente variabilidad (menos de 2 categorías en una de las variables).")
        p_val = np.nan # Asigna NaN si el test no se puede realizar

    # 3. Calcular la tasa de incumplimiento por categoría
    # Usamos .mean() ya que Incumplimiento es 0/1, la media es la proporción de 1s.
    incumplimiento_rate = df.groupby(col)[target_variable].mean() * 100
    print("\nTasa de Incumplimiento (%) promedio por categoría:")
    print(incumplimiento_rate)

    # Guardar resultados
    # Es más flexible guardar todas las categorías en un diccionario por variable.
    incumplimiento_rates_dict = incumplimiento_rate.to_dict()
    results_categorical_entry = {'Variable': col, 'Chi2_p_value': p_val}
    for category, rate in incumplimiento_rates_dict.items():
        results_categorical_entry[f'Tasa_Incumplimiento_Cat_{category}'] = rate
    results_categorical.append(results_categorical_entry)

    # 4. Visualización: Gráfico de barras de la tasa de incumplimiento
    plt.figure(figsize=(8, 5))
    sns.barplot(x=df[col], y=df[target_variable], palette='viridis', ci=None)
    plt.title(f'Tasa de Incumplimiento por {col}')
    plt.xlabel(col)
    plt.ylabel('Tasa de Incumplimiento')
    plt.xticks(rotation=45, ha='right') # Rotar etiquetas para mejor legibilidad
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout() # Ajustar el diseño para evitar superposiciones
    plt.show()

# Resumen de resultados para variables categóricas
results_df_categorical = pd.DataFrame(results_categorical)
print("\n" + "="*70)
print("RESUMEN DEL ANÁLISIS BIVARIADO (Variables Categóricas)")
print("="*70 + "\n")
print(results_df_categorical)



## Análisis para Variables Numéricas


results_numeric = []

print("\n" + "="*70)
print("INICIANDO ANÁLISIS BIVARIADO: VARIABLES NUMÉRICAS vs. INCUMPLIMIENTO")
print("="*70 + "\n")

for col in numeric_cols:
    print(f"\n--- Analizando: {col} ---")

    # 1. Estadísticas descriptivas de la variable numérica agrupada por 'Incumplimiento'
    print("\nEstadísticas descriptivas por grupo de Incumplimiento:")
    print(df.groupby(target_variable)[col].describe())

    # 2. Test t de Student para la diferencia de medias
    # Se recomienda el test t de Welch (equal_var=False) si no se asume igualdad de varianzas.
    group_0 = df[df[target_variable] == 0][col].dropna()
    group_1 = df[df[target_variable] == 1][col].dropna()

    p_val_ttest = np.nan # Inicializar p_val para el caso de no tener suficientes datos

    if len(group_0) > 1 and len(group_1) > 1: # Asegúrate de tener suficientes datos para el test
        ttest_stat, p_val_ttest = ttest_ind(group_0, group_1, equal_var=False) # Welch's t-test
        print(f"\nTest t de Student (Welch's) - p-value: {p_val_ttest:.4f}")
        if p_val_ttest < 0.05:
            print(f"  -> Hay una diferencia estadísticamente significativa en las medias de {col} entre los grupos de {target_variable} (p < 0.05).")
        else:
            print(f"  -> No hay evidencia de una diferencia estadísticamente significativa en las medias de {col} entre los grupos de {target_variable} (p >= 0.05).")
    else:
        print(f"  -> No hay suficientes datos en uno o ambos grupos para realizar el test t para {col}.")

    results_numeric.append({
        'Variable': col,
        'Media_No_Incumplimiento': df[df[target_variable] == 0][col].mean(),
        'Media_Incumplimiento': df[df[target_variable] == 1][col].mean(),
        'T_Test_p_value': p_val_ttest
    })

    # 3. Visualización: Boxplot
    plt.figure(figsize=(8, 5))
    sns.boxplot(x=df[target_variable], y=df[col], palette='coolwarm')
    plt.title(f'Distribución de {col} por Grupo de Incumplimiento')
    plt.xlabel(f'{target_variable} (0: No Incumplimiento, 1: Incumplimiento)')
    plt.ylabel(col)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

    # 4. Visualización: KDE Plot (Densidad)
    plt.figure(figsize=(10, 6))
    sns.kdeplot(df[df[target_variable] == 0][col].dropna(), label='No Incumplimiento (0)', fill=True, alpha=0.5, linewidth=2)
    sns.kdeplot(df[df[target_variable] == 1][col].dropna(), label='Incumplimiento (1)', fill=True, alpha=0.5, linewidth=2)
    plt.title(f'Distribución de {col} por Grupo de Incumplimiento')
    plt.xlabel(col)
    plt.ylabel('Densidad')
    plt.legend()
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# Resumen de resultados para variables numéricas
results_df_numeric = pd.DataFrame(results_numeric)
print("\n" + "="*70)
print("RESUMEN DEL ANÁLISIS BIVARIADO (Variables Numéricas)")
print("="*70 + "\n")
print(results_df_numeric)

print("\nAnálisis bivariado completado. Revisa los resultados y las visualizaciones para identificar las variables más influyentes.")

plt.figure(figsize=(10,6))
sns.set_style('whitegrid')
plt.hist(df['Indice_Incumplimiento'], bins=30,alpha=0.7,color='skyblue',edgecolor='black')
plt.title('Histograma de frecuencias',fontsize=15)
plt.xlabel('valores',fontsize=12)
plt.ylabel('Frecuencias',fontsize=12)
plt.grid(axis='y', alpha=0.7)

plt.figure(figsize=(10,6))
sns.set_style('whitegrid')

# Crear el gráfico de conteo
ax = sns.countplot(x='Incumplimiento', data=df, color='skyblue', edgecolor='black')

# Añadir las etiquetas de los conteos en las barras
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 5), textcoords='offset points', fontsize=10)

plt.title('Distribución de la Variable Objetivo "Incumplimiento"', fontsize=15)
plt.xlabel('Incumplimiento (0: No Incumple, 1: Incumple)', fontsize=12) # Ajusta la etiqueta si 0/1 significan otra cosa
plt.ylabel('Frecuencia', fontsize=12)
plt.grid(axis='y', alpha=0.7)
plt.show()

from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    confusion_matrix,
    classification_report,
    roc_curve,
    auc
)

from imblearn.over_sampling import SMOTE
from collections import Counter # Para verificar el balanceo

# Configuración para gráficos
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['font.size'] = 12

# --- 0. Pre-requisitos: Asumimos que tu DataFrame 'df' está listo ---
# 'df' debe contener todas las columnas procesadas (dummies, numéricas, etc.)
# y la variable objetivo 'Incumplimiento'.
# ¡Es CRÍTICO que tu DataFrame 'df' ya no tenga columnas duplicadas o las originales
# que se usaron para crear las dummies o el índice!
# Las columnas deben ser las que listaste anteriormente después del one-hot encoding:
# Incumplimiento, Tipo_Empleo_Empleado, Tipo_Empleo_Estudiante, Tipo_Empleo_Independiente,
# Estado_Civil_Divorciado, Estado_Civil_Soltero, Estado_Civil_Viudo, Sexo_Masculino,
# Indice_Incumplimiento, Ratio_Deuda_Ingreso,
# Edad_18-30, Edad_31-45, Edad_46-60, Edad_61-75, Edad_76+,
# Historial_Bajo, Historial_Regular, Historial_Bueno, Historial_Excelente,
# Ingreso_Bajo, Ingreso_Medio, Ingreso_Alto,
# Deuda_Bajo, Deuda_Medio, Deuda_Alto,
# Tiempo_1 a 12 meses, Tiempo_1 a 5 años, Tiempo_5 a 10 años, Tiempo_Desempleado, Tiempo_Más de 10 años,
# Plazo_Corto_Plazo, Plazo_Mediano_Plazo, Plazo_Largo_Plazo


print("--- 1. Preparación de Datos ---")
print(f"Dimensiones del DataFrame inicial: {df.shape}")
print(f"Balance de la variable objetivo 'Incumplimiento' antes de SMOTE:\n{df['Incumplimiento'].value_counts(normalize=True)}")

# 1.1 Separar características (X) y variable objetivo (y)
X = df.drop('Incumplimiento', axis=1)
y = df['Incumplimiento']

# 1.2 Identificar columnas numéricas para escalado
# Se asume que las columnas booleanas (Tipo_Empleo_*, Estado_Civil_*, Sexo_Masculino)
# ya se convirtieron a int (0/1) y no necesitan escalado.
# Las dummies (Edad_*, Historial_*, etc.) tampoco necesitan escalado.
numeric_features = ['Indice_Incumplimiento', 'Ratio_Deuda_Ingreso']

# 1.3 Dividir los datos en conjuntos de entrenamiento y prueba (antes de SMOTE)
# Se usa 'stratify=y' para mantener la proporción de clases en ambos conjuntos.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

print(f"\nDimensiones de X_train: {X_train.shape}, y_train: {y_train.shape}")
print(f"Dimensiones de X_test: {X_test.shape}, y_test: {y_test.shape}")
print(f"Balance de y_train antes de SMOTE:\n{y_train.value_counts(normalize=True)}")

# 1.4 Escalado de características numéricas (solo en el conjunto de entrenamiento, luego se aplica al de prueba)
scaler = StandardScaler()
X_train[numeric_features] = scaler.fit_transform(X_train[numeric_features])
X_test[numeric_features] = scaler.transform(X_test[numeric_features]) # Aplicar la misma transformación al conjunto de prueba

print("\nCaracterísticas numéricas escaladas.")

# 1.5 Balanceo de clases con SMOTE (solo en el conjunto de entrenamiento)
print("\n--- 2. Balanceo de Clases con SMOTE ---")
print(f"Balance de clases en y_train ANTES de SMOTE: {Counter(y_train)}")

smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)

print(f"Balance de clases en y_train DESPUÉS de SMOTE: {Counter(y_train_res)}")
print(f"Dimensiones de X_train_res: {X_train_res.shape}, y_train_res: {y_train_res.shape}")

# --- 3. Entrenamiento y Evaluación de Modelos ---
print("\n--- 3. Entrenamiento y Evaluación de Modelos ---")

models = {
    'Regresión Logística': LogisticRegression(random_state=42, solver='liblinear'), # liblinear es bueno para datasets pequeños
    'Random Forest': RandomForestClassifier(random_state=42, n_estimators=100),
    'XGBoost': xgb.XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss')
}

results = {}

for name, model in models.items():
    print(f"\nEntrenando y evaluando: {name}...")

    # Entrenar el modelo con los datos balanceados
    model.fit(X_train_res, y_train_res)

    # Predicciones en el conjunto de PRUEBA (NO balanceado)
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1] # Probabilidades para la clase positiva

    # Calcular métricas
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, zero_division=0) # zero_division=0 para evitar warnings si no hay predicciones
    recall = recall_score(y_test, y_pred, zero_division=0)
    f1 = f1_score(y_test, y_pred, zero_division=0)
    roc_auc = roc_auc_score(y_test, y_proba)
    cm = confusion_matrix(y_test, y_pred)

    results[name] = {
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1-Score': f1,
        'ROC-AUC': roc_auc,
        'Confusion Matrix': cm,
        'y_pred': y_pred,
        'y_proba': y_proba
    }

    print(f"  Accuracy: {accuracy:.4f}")
    print(f"  Precision: {precision:.4f}")
    print(f"  Recall (Sensibilidad): {recall:.4f}")
    print(f"  F1-Score: {f1:.4f}")
    print(f"  ROC-AUC: {roc_auc:.4f}")
    print(f"  Matriz de Confusión:\n{cm}")
    print(f"  Reporte de Clasificación:\n{classification_report(y_test, y_pred, zero_division=0)}")

# --- 4. Comparación de Resultados y Gráficos ---
print("\n--- 4. Comparación de Resultados y Gráficos ---")

# Dataframe para comparar métricas
metrics_df = pd.DataFrame({
    'Modelo': list(results.keys()),
    'Accuracy': [res['Accuracy'] for res in results.values()],
    'Precision': [res['Precision'] for res in results.values()],
    'Recall': [res['Recall'] for res in results.values()],
    'F1-Score': [res['F1-Score'] for res in results.values()],
    'ROC-AUC': [res['ROC-AUC'] for res in results.values()]
})

print("\nTabla Comparativa de Métricas:")
print(metrics_df.round(4).to_string(index=False))

# Gráfico de barras comparativo de métricas
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
fig.suptitle('Comparación de Métricas de Rendimiento del Modelo', fontsize=16)

metrics_to_plot = ['Accuracy', 'Precision', 'Recall', 'F1-Score', 'ROC-AUC']
ax_flat = axes.flatten()

for i, metric in enumerate(metrics_to_plot):
    sns.barplot(x='Modelo', y=metric, data=metrics_df, ax=ax_flat[i], palette='viridis')
    ax_flat[i].set_title(f'{metric} por Modelo')
    ax_flat[i].set_ylim(0, 1)
    # Corrección para la rotación y alineación de las etiquetas del eje x
    ax_flat[i].set_xticklabels(ax_flat[i].get_xticklabels(), rotation=30, ha='right')
    for p in ax_flat[i].patches:
        ax_flat[i].annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                            ha='center', va='center', xytext=(0, 5), textcoords='offset points', fontsize=10)

# Dejar el último subplot vacío o usarlo para algo más si se desea.
fig.delaxes(ax_flat[5]) # Si solo tienes 5 métricas y 6 subplots

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

# Gráfico de Curvas ROC
plt.figure(figsize=(10, 8))
for name, res in results.items():
    fpr, tpr, _ = roc_curve(y_test, res['y_proba'])
    roc_auc_val = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc_val:.4f})')

plt.plot([0, 1], [0, 1], 'k--', label='Clasificador Aleatorio (AUC = 0.50)')
plt.xlabel('Tasa de Falsos Positivos (FPR)')
plt.ylabel('Tasa de Verdaderos Positivos (TPR) / Recall')
plt.title('Curvas ROC para Modelos de Clasificación')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()

# Gráfico de Matrices de Confusión
fig, axes = plt.subplots(1, len(models), figsize=(5 * len(models), 5))
if len(models) == 1: # Si solo hay un modelo, axes no es un array
    axes = [axes]
fig.suptitle('Matrices de Confusión', fontsize=16)

for i, (name, res) in enumerate(results.items()):
    sns.heatmap(res['Confusion Matrix'], annot=True, fmt='d', cmap='Blues', ax=axes[i], cbar=False)
    axes[i].set_title(f'{name}')
    axes[i].set_xlabel('Predicho')
    axes[i].set_ylabel('Real')
    axes[i].set_ylim(2, 0) # Ajuste para que las etiquetas no se corten en algunas versiones de matplotlib

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

print("\n--- Conclusión ---")
print("Para tu objetivo de identificar a solicitantes con alta probabilidad de incumplimiento y reducir pérdidas,")
print("la métrica 'Recall' para la clase minoritaria (1: Incumplimiento) es crucial, junto con el 'F1-Score' y 'ROC-AUC'.")
print("Un alto Recall significa que el modelo es bueno capturando a la mayoría de los incumplidores reales.")
print("Analiza la tabla de métricas y los gráficos para determinar qué modelo se alinea mejor con tus objetivos de negocio.")


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    confusion_matrix,
    classification_report,
    roc_curve,
    auc,
    PrecisionRecallDisplay # Para la curva Precision-Recall
)

from imblearn.over_sampling import SMOTE
from collections import Counter

# Configuración para gráficos
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['font.size'] = 12


# --- 1. Optimización de Hiperparámetros con GridSearchCV ---
print("\n--- 1. Optimización de Hiperparámetros con GridSearchCV ---")

# Definir el modelo base
log_reg = LogisticRegression(random_state=42, solver='liblinear')

# Definir la cuadrícula de hiperparámetros a buscar
# 'C': Inverso de la fuerza de regularización; valores más pequeños indican mayor regularización.
#      La regularización ayuda a prevenir el overfitting.
# 'penalty': Tipo de regularización ('l1' o 'l2'). 'l1' puede llevar a la selección de características.
param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'penalty': ['l1', 'l2']
}

# Configurar GridSearchCV
# cv=StratifiedKFold: Esencial para mantener las proporciones de clase en cada fold de validación cruzada.
# scoring='recall': Optimizamos directamente para el recall de la clase positiva (1, Incumplimiento),
#                    que es nuestra métrica principal de negocio.
# verbose=1: Muestra el progreso.
# n_jobs=-1: Utiliza todos los núcleos de CPU disponibles para acelerar la búsqueda.
grid_search = GridSearchCV(
    estimator=log_reg,
    param_grid=param_grid,
    cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=42), # 5-fold cross-validation
    scoring='recall', # Optimizar para Recall de la clase positiva
    verbose=1,
    n_jobs=-1
)

# Ejecutar la búsqueda en los datos de entrenamiento balanceados
print("Realizando la búsqueda de hiperparámetros (esto puede tardar unos minutos)...")
grid_search.fit(X_train_res, y_train_res)

# Obtener los mejores hiperparámetros y el mejor score
best_params = grid_search.best_params_
best_score = grid_search.best_score_
print(f"\nMejores Hiperparámetros encontrados: {best_params}")
print(f"Mejor Recall promedio en CV con estos parámetros: {best_score:.4f}")

# Obtener el mejor modelo
best_log_reg_model = grid_search.best_estimator_

# --- 2. Evaluación del Modelo Optimizado en el Conjunto de Prueba ---
print("\n--- 2. Evaluación del Modelo Optimizado ---")

y_pred_opt = best_log_reg_model.predict(X_test)
y_proba_opt = best_log_reg_model.predict_proba(X_test)[:, 1] # Probabilidades para la clase positiva

# Calcular y mostrar métricas para el modelo optimizado (usando umbral por defecto 0.5)
accuracy_opt = accuracy_score(y_test, y_pred_opt)
precision_opt = precision_score(y_test, y_pred_opt, zero_division=0)
recall_opt = recall_score(y_test, y_pred_opt, zero_division=0)
f1_opt = f1_score(y_test, y_pred_opt, zero_division=0)
roc_auc_opt = roc_auc_score(y_test, y_proba_opt)
cm_opt = confusion_matrix(y_test, y_pred_opt)

print(f"Rendimiento del Modelo Optimizado (Umbral por defecto 0.5):")
print(f"  Accuracy: {accuracy_opt:.4f}")
print(f"  Precision: {precision_opt:.4f}")
print(f"  Recall (Sensibilidad): {recall_opt:.4f}")
print(f"  F1-Score: {f1_opt:.4f}")
print(f"  ROC-AUC: {roc_auc_opt:.4f}")
print(f"  Matriz de Confusión:\n{cm_opt}")
print(f"  Reporte de Clasificación:\n{classification_report(y_test, y_pred_opt, zero_division=0)}")

# --- 3. Ajuste del Umbral de Decisión ---
print("\n--- 3. Ajuste del Umbral de Decisión ---")

# La Regresión Logística devuelve probabilidades. Un umbral de 0.5 significa:
# Si P(Incumplimiento) > 0.5, predice 1 (Incumplimiento)
# Si P(Incumplimiento) <= 0.5, predice 0 (No Incumplimiento)

# Queremos explorar cómo diferentes umbrales impactan Precision, Recall y F1-Score.
# La clase 1 (Incumplimiento) es nuestra clase de interés principal.
# Si queremos aumentar el Recall (identificar más incumplidores), generalmente debemos bajar el umbral.
# Esto significa que seremos más "sensibles" y clasificaremos más casos como incumplimiento,
# pero podría aumentar los falsos positivos (disminuir la precisión).

thresholds = np.arange(0.1, 0.9, 0.05) # Umbrales a probar, de 0.1 a 0.85 en pasos de 0.05
threshold_metrics = []

for t in thresholds:
    # Predicciones con el nuevo umbral
    y_pred_threshold = (y_proba_opt >= t).astype(int)

    # Calcular métricas para la clase 1 (Incumplimiento)
    precision_t = precision_score(y_test, y_pred_threshold, pos_label=1, zero_division=0)
    recall_t = recall_score(y_test, y_pred_threshold, pos_label=1, zero_division=0)
    f1_t = f1_score(y_test, y_pred_threshold, pos_label=1, zero_division=0)

    # También, podemos monitorear los Falsos Negativos si quieres, directamente desde la matriz
    cm_t = confusion_matrix(y_test, y_pred_threshold)
    # Asegúrate de que cm_t tenga las dimensiones esperadas (2x2)
    # y que los índices de las clases 0 y 1 sean consistentes.
    # Esto asume: cm = [[TN, FP], [FN, TP]]
    fn_t = cm_t[1, 0] if cm_t.shape == (2,2) else -1 # Falsos Negativos
    fp_t = cm_t[0, 1] if cm_t.shape == (2,2) else -1 # Falsos Positivos

    threshold_metrics.append({
        'Umbral': t,
        'Precision_1': precision_t,
        'Recall_1': recall_t,
        'F1_1': f1_t,
        'Falsos_Negativos_1': fn_t,
        'Falsos_Positivos_1': fp_t
    })

threshold_df = pd.DataFrame(threshold_metrics)
print("\nTabla de Métricas vs. Umbral de Decisión:")
print(threshold_df.round(4).to_string(index=False))

# --- 4. Visualización del Ajuste del Umbral ---
print("\n--- 4. Visualización del Ajuste del Umbral ---")

# Gráfico de Precision-Recall Trade-off
plt.figure(figsize=(10, 7))
plt.plot(threshold_df['Umbral'], threshold_df['Precision_1'], label='Precision (Clase 1)')
plt.plot(threshold_df['Umbral'], threshold_df['Recall_1'], label='Recall (Clase 1)')
plt.plot(threshold_df['Umbral'], threshold_df['F1_1'], label='F1-Score (Clase 1)', linestyle='--', marker='o', markersize=4)
plt.axvline(x=0.5, color='r', linestyle=':', label='Umbral por Defecto (0.5)')
plt.xlabel('Umbral de Probabilidad')
plt.ylabel('Valor de la Métrica')
plt.title('Precision, Recall y F1-Score vs. Umbral de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()

# Curva Precision-Recall (alternativa para visualizar el trade-off)
# Muestra la precisión para varios umbrales de recall.
plt.figure(figsize=(10, 7))
display = PrecisionRecallDisplay.from_estimator(
    best_log_reg_model, X_test, y_test, name="Regresión Logística Optimizada"
)
display.ax_.set_title("Curva Precision-Recall")
plt.show()

# --- 5. Elección del Umbral Óptimo (basado en el objetivo de negocio) ---
print("\n--- 5. Elección del Umbral Óptimo ---")
print("Para elegir el umbral óptimo, debes considerar el equilibrio entre:")
print("- Minimizar Falsos Negativos (maximizar Recall de la Clase 1): Directamente reduce las pérdidas por morosidad.")
print("- Aceptar un nivel razonable de Falsos Positivos (mantener una Precisión aceptable para la Clase 1): Evitar denegar excesivamente créditos a buenos clientes.")
print("Analiza la tabla 'Métricas vs. Umbral' y el gráfico para identificar el umbral que mejor se alinee con las políticas de riesgo y costos del banco.")

# Ejemplo: Elegir un umbral que dé un recall del 70% o más para la clase 1,
# mientras se mantiene una precisión aceptable.
# O elegir el umbral donde el F1-Score es más alto, que es un buen compromiso.

# Puedes elegir un umbral específico después de revisar la tabla.
# Por ejemplo, si el objetivo es un recall de al menos 0.75 y ves un umbral
# que lo cumple con una precisión razonable, ese podría ser tu elección.
chosen_threshold = 0.5 # Ejemplo: podrías cambiarlo a 0.4 o 0.45 si te dan mejor recall sin sacrificar mucho precision

# Re-evaluar el modelo con el umbral elegido (si es diferente de 0.5)
if chosen_threshold != 0.5:
    y_pred_final = (y_proba_opt >= chosen_threshold).astype(int)
    print(f"\n--- Rendimiento del Modelo con Umbral Elegido ({chosen_threshold:.2f}) ---")
    print(f"Reporte de Clasificación Final:\n{classification_report(y_test, y_pred_final, zero_division=0)}")
    print(f"Matriz de Confusión Final:\n{confusion_matrix(y_test, y_pred_final)}")

# Configuración para gráficos
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['font.size'] = 12


# --- 1. Optimización de Hiperparámetros con GridSearchCV ---
print("\n--- 1. Optimización de Hiperparámetros con GridSearchCV ---")

# Definir el modelo base
log_reg = LogisticRegression(random_state=42, solver='liblinear')

# Definir la cuadrícula de hiperparámetros a buscar
param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'penalty': ['l1', 'l2']
}

# Configurar GridSearchCV
grid_search = GridSearchCV(
    estimator=log_reg,
    param_grid=param_grid,
    cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=42), # 5-fold cross-validation
    scoring='recall', # Optimizar para Recall de la clase positiva
    verbose=1,
    n_jobs=-1
)

# Ejecutar la búsqueda en los datos de entrenamiento balanceados
print("Realizando la búsqueda de hiperparámetros (esto puede tardar unos minutos)...")
grid_search.fit(X_train_res, y_train_res)

# Obtener los mejores hiperparámetros y el mejor score
best_params = grid_search.best_params_
best_score = grid_search.best_score_
print(f"\nMejores Hiperparámetros encontrados: {best_params}")
print(f"Mejor Recall promedio en CV con estos parámetros: {best_score:.4f}")

# Obtener el mejor modelo
best_log_reg_model = grid_search.best_estimator_

# --- 2. Evaluación del Modelo Optimizado en el Conjunto de Prueba (Umbral 0.5 por defecto) ---
print("\n--- 2. Evaluación del Modelo Optimizado (Umbral por defecto 0.5) ---")

y_pred_opt = best_log_reg_model.predict(X_test)
y_proba_opt = best_log_reg_model.predict_proba(X_test)[:, 1] # Probabilidades para la clase positiva

# Calcular y mostrar métricas para el modelo optimizado
accuracy_opt = accuracy_score(y_test, y_pred_opt)
precision_opt = precision_score(y_test, y_pred_opt, zero_division=0)
recall_opt = recall_score(y_test, y_pred_opt, zero_division=0)
f1_opt = f1_score(y_test, y_pred_opt, zero_division=0)
roc_auc_opt = roc_auc_score(y_test, y_proba_opt)
cm_opt = confusion_matrix(y_test, y_pred_opt)

print(f"  Accuracy: {accuracy_opt:.4f}")
print(f"  Precision (Clase 1): {precision_opt:.4f}")
print(f"  Recall (Clase 1): {recall_opt:.4f}")
print(f"  F1-Score (Clase 1): {f1_opt:.4f}")
print(f"  ROC-AUC: {roc_auc_opt:.4f}")
print(f"  Matriz de Confusión:\n{cm_opt}")
print(f"  Reporte de Clasificación:\n{classification_report(y_test, y_pred_opt, zero_division=0)}")


# --- 3. Ajuste del Umbral de Decisión y Visualización ---
print("\n--- 3. Ajuste del Umbral de Decisión y Visualización ---")

# Umbrales a probar
thresholds = np.arange(0.1, 0.9, 0.05)
threshold_metrics = []

for t in thresholds:
    y_pred_threshold = (y_proba_opt >= t).astype(int)

    precision_t = precision_score(y_test, y_pred_threshold, pos_label=1, zero_division=0)
    recall_t = recall_score(y_test, y_pred_threshold, pos_label=1, zero_division=0)
    f1_t = f1_score(y_test, y_pred_threshold, pos_label=1, zero_division=0)

    cm_t = confusion_matrix(y_test, y_pred_threshold)
    fn_t = cm_t[1, 0] if cm_t.shape == (2,2) else -1
    fp_t = cm_t[0, 1] if cm_t.shape == (2,2) else -1

    threshold_metrics.append({
        'Umbral': t,
        'Precision_1': precision_t,
        'Recall_1': recall_t,
        'F1_1': f1_t,
        'Falsos_Negativos_1': fn_t,
        'Falsos_Positivos_1': fp_t
    })

threshold_df = pd.DataFrame(threshold_metrics)
print("\nTabla de Métricas vs. Umbral de Decisión:")
print(threshold_df.round(4).to_string(index=False))

# Gráfico de Precision-Recall Trade-off
plt.figure(figsize=(10, 7))
plt.plot(threshold_df['Umbral'], threshold_df['Precision_1'], label='Precision (Clase 1)')
plt.plot(threshold_df['Umbral'], threshold_df['Recall_1'], label='Recall (Clase 1)')
plt.plot(threshold_df['Umbral'], threshold_df['F1_1'], label='F1-Score (Clase 1)', linestyle='--', marker='o', markersize=4)
plt.axvline(x=0.5, color='r', linestyle=':', label='Umbral por Defecto (0.5)')
plt.xlabel('Umbral de Probabilidad')
plt.ylabel('Valor de la Métrica')
plt.title('Precision, Recall y F1-Score vs. Umbral de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()

# Curva Precision-Recall
plt.figure(figsize=(10, 7))
display = PrecisionRecallDisplay.from_estimator(
    best_log_reg_model, X_test, y_test, name="Regresión Logística Optimizada"
)
display.ax_.set_title("Curva Precision-Recall")
plt.show()

# --- 4. Elección y Re-evaluación con el Umbral Final Elegido ---
print("\n--- 4. Elección y Re-evaluación con el Umbral Final Elegido ---")

# Definimos el umbral final elegido
chosen_threshold = 0.40 # ajuste del umbral establecido en el paso anterior

# Generar predicciones con el umbral elegido
y_pred_final = (y_proba_opt >= chosen_threshold).astype(int)

# Calcular y mostrar métricas finales con el umbral elegido
print(f"Rendimiento del Modelo de Regresión Logística con Umbral {chosen_threshold:.2f}:")
final_accuracy = accuracy_score(y_test, y_pred_final)
final_precision = precision_score(y_test, y_pred_final, zero_division=0)
final_recall = recall_score(y_test, y_pred_final, zero_division=0)
final_f1 = f1_score(y_test, y_pred_final, zero_division=0)
final_roc_auc = roc_auc_score(y_test, y_proba_opt) # AUC no cambia con el umbral

print(f"  Accuracy: {final_accuracy:.4f}")
print(f"  Precision (Clase 1): {final_precision:.4f}")
print(f"  Recall (Clase 1): {final_recall:.4f}")
print(f"  F1-Score (Clase 1): {final_f1:.4f}")
print(f"  ROC-AUC: {final_roc_auc:.4f}") # ROC-AUC no se ve afectado por el umbral

print(f"\n  Matriz de Confusión Final (Umbral {chosen_threshold:.2f}):")
final_cm = confusion_matrix(y_test, y_pred_final)
print(final_cm)

print(f"\n  Reporte de Clasificación Final (Umbral {chosen_threshold:.2f}):")
print(classification_report(y_test, y_pred_final, zero_division=0))

# --- 5. Análisis del Impacto del Umbral (Opcional, para la discusión) ---
print("\n--- 5. Análisis del Impacto del Umbral (Resumen) ---")
print(f"Al cambiar el umbral de 0.50 a {chosen_threshold:.2f} para la clase 1 (Incumplimiento):")
print(f"  - El **Recall (Clase 1)** mejoró de {recall_opt:.4f} a {final_recall:.4f}.")
print(f"    Esto significa que el modelo ahora identifica a más incumplidores reales, reduciendo las pérdidas por morosidad.")
print(f"  - La **Precision (Clase 1)** pasó de {precision_opt:.4f} a {final_precision:.4f}.")
print(f"    Esto implica que, de las veces que el modelo predice incumplimiento, una menor proporción es correcta.")
print(f"  - Los **Falsos Negativos (Clase 1)** se redujeron de {cm_opt[1,0]} a {final_cm[1,0]}.")
print(f"  - Los **Falsos Positivos (Clase 0)** aumentaron de {cm_opt[0,1]} a {final_cm[0,1]}.")
print("Este trade-off entre Recall y Precision es una decisión estratégica de negocio. El umbral de 0.40 representa un equilibrio.")
```

```python
# archivo: pipeline_model.py

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, roc_auc_score
from sklearn.impute import SimpleImputer
from joblib import dump

# 1. Cargar datos
data = pd.read_csv("data_credito.csv")

# 2. Eliminar columnas no útiles
data.drop(columns=['ID_Cliente'], inplace=True)

# 3. Definir X e y
X = data.drop('Incumplimiento', axis=1)
y = data['Incumplimiento']

# 4. Identificar columnas
numeric_features = X.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_features = X.select_dtypes(include=['object', 'category']).columns.tolist()

# 5. Preprocesamiento
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(drop='first', handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ]
)

# 6. Pipeline completo
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression(max_iter=1000, random_state=42))
])

# 7. Entrenar modelo
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)
pipeline.fit(X_train, y_train)

# 8. Evaluación
y_pred = pipeline.predict(X_test)
y_prob = pipeline.predict_proba(X_test)[:, 1]
print(classification_report(y_test, y_pred))
print("AUC-ROC:", roc_auc_score(y_test, y_prob))

# 9. Guardar pipeline entrenado
dump(pipeline, "best_log_reg_model.joblib")
```
